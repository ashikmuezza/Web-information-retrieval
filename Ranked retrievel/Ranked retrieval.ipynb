```python
D1= "shall see sun soon"
D2= "we shall see sun shine soon"

def intersection(a,b):
  return set(set(a.split()) & set(b.split()))

def union(a,b):
  return set(a.split() + b.split())


jaccard_similarity = len(intersection(D1,D2)) / len(union(D1,D2))

print('jaccard_similarity ',jaccard_similarity)
```
